```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("heart.csv")
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['age'], bins=15, color='skyblue', edgecolor='black')
plt.title("Age Distribution of Patients")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
labels = ['No Disease', 'Heart Disease']
sizes = df['target'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=['lightgreen', 'salmon'])
plt.title("Heart Disease Presence Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='target', y='chol', data=df)
plt.title("Cholesterol Levels by Heart Disease Status")
plt.xlabel("Heart Disease (0 = No, 1 = Yes)")
plt.ylabel("Cholesterol")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='age', y='thalach', hue='target', data=df)
plt.title("Age vs Max Heart Rate (thalach)")
plt.xlabel("Age")
plt.ylabel("Max Heart Rate")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='sex', y='thalach', data=df, palette='pastel')
sns.stripplot(x='sex', y='thalach', data=df, color='black', alpha=0.5, jitter=True)
plt.title("Max Heart Rate by Sex with Distribution")
plt.xlabel("Sex (0 = Female, 1 = Male)")
plt.ylabel("Max Heart Rate")
plt.show()
```
