```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Load the dataset
heart_df = pd.read_csv('heart.csv')

# Display first few rows
heart_df.head()
```

```python
sns.set(style="whitegrid")

# a) Histogram
plt.figure(figsize=(8,5))
plt.hist(heart_df['age'], bins=15, edgecolor='black')
plt.title('Age Distribution of Patients')
plt.xlabel('Age')
plt.ylabel('Number of Patients')
plt.grid(True)
plt.show()
```

```python
# b) Dot Plot
plt.figure(figsize=(10,4))
plt.plot(heart_df['age'], 'o', markersize=5)
plt.title('Dot Plot of Patient Ages')
plt.xlabel('Patient Index')
plt.ylabel('Age')
plt.grid(True)
plt.show()

# Objective: Visualize **individual patient ages** to detect spread, clusters, or gaps across the dataset.
```

```python
# c) Bar Plot
plt.figure(figsize=(7,5))
sns.barplot(x='target', y='chol', data=heart_df, ci=None)
plt.title('Average Cholesterol Level by Heart Disease Presence')
plt.xlabel('Heart Disease (0 = No, 1 = Yes)')
plt.ylabel('Average Cholesterol Level')
plt.show()

# Objective: Compare **average cholesterol levels** between patients **with and without heart disease**.
```

```python
# d) Line Chart
heart_df_sorted = heart_df.sort_values('age')

plt.figure(figsize=(10,5))
plt.plot(heart_df_sorted['age'], heart_df_sorted['thalach'], marker='o', linestyle='-')
plt.title('Trend of Maximum Heart Rate (thalach) vs Age')
plt.xlabel('Age')
plt.ylabel('Maximum Heart Rate Achieved (thalach)')
plt.grid(True)
plt.show()

# Objective: Observe **how maximum heart rate (thalach)** varies as **patient age increases**.
```

```python
# e) Box Plot with Histogram and Scatter Plot Overlay

plt.figure(figsize=(12,8))

# Box plot for age
sns.boxplot(x=heart_df['age'], color='lightblue', width=0.3)

# Scatter plot (dot plot) overlay
sns.stripplot(x=heart_df['age'], color='red', size=5, jitter=True)

# Histogram overlay (using kdeplot for smoothness)
sns.histplot(heart_df['age'], bins=15, color='lightgreen', alpha=0.3)

plt.title('Box Plot with Scatter and Histogram Overlay (Age)')
plt.xlabel('Age')
plt.show()

# Objective: Analyze **distribution (histogram)**, **outliers (box plot)**, and **individual data points (scatter)** together in one visual for Age.
```
