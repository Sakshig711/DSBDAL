```python
import pandas as pd
```

```python
df = pd.read_csv('dataset_Facebook.csv', sep=';')
```

```python
df.head()
```

```python
# 2a. Create subsets by “Type”
photo = df[df["Type"] == "Photo"] #only rows with type photo
video = df[df["Type"] == "Video"] #only rows with type video
```

```python
print("Photo subset:", photo.shape)
print("Video subset:", video.shape)
```

```python
photo = photo.copy()
video = video.copy()

photo['Type'] = photo['Type'].str.strip().str.lower()
video['Type'] = video['Type'].str.strip().str.lower()

merged = pd.merge(photo, video, on='Type', how='inner')
#pd.concat([photo, video], axis=0)
```

```python
print("Merged subset:",  merged.shape)
```

```python
df_sorted = df.sort_values(by='Page total likes', ascending=False)
print("\nTop 6 by likes:\n", df_sorted[['Type','Page total likes']].head(6))
```
